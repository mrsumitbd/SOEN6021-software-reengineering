```python
import ssl
ssl._create_default_https_context = ssl._create_unverified_context
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
df = pd.read_csv('pct_change_exp_os.csv')
df['rnd'] = [''] * df.shape[0]
#df.rename(columns={'OS': 'Operating System'}, inplace=True)

df.head(100)
```

```python
sns.violinplot(data = df, x = 'rnd', y = 'pct_change', hue = 'Operating System', split = True);
plt.tick_params(bottom = False)
plt.xlabel("")
plt.ylabel("Percentage Difference")
#plt.savefig('pct_diff_exp_py.png', dpi = 300);
```

```python
sns.boxplot(data =df.loc[df['CPU Architecture'] == 'ARM64'], y = 'pct_change', x = 'rnd', orient='v' )
plt.tick_params(bottom = False)
plt.xlabel("")
plt.ylabel("Percentage Difference")
plt.savefig('pct_diff_exp_hw.png', dpi = 300);
```
